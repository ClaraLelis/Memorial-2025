<font color = green> Neste caderno é feito um exemplo de algoritmo genético feito para adivinhar uma senha cujo tamanho é desconhecido. Para tal utilizou-se as funções presentes no script `fera_9` que são melhor explicadas ao final do caderno.

```python
from fera_9 import *
from string import ascii_lowercase, ascii_uppercase, digits
```

```python
TARGET="Jamiel2025" 
CHARS=ascii_lowercase+ascii_uppercase+digits
POP_SIZE=100
MUT_RATE=1
MAX_GEN=1000

pop=[create_individual(CHARS) for _ in range(POP_SIZE)]

for gen in range(MAX_GEN):
    pop=sorted(pop, key=lambda ind: fitness(ind, TARGET), reverse=True)
    best=pop[0]
    print(f"{gen}: {best}")

    if best==TARGET:
        print(f"Senha encontrada na geração {gen}: {best}")
        break
    next_gen=pop[:10]

    while len(next_gen)<POP_SIZE:
        parent1, parent2=random.choices(pop[:50], k=2)
        child=crossover(parent1, parent2)
        if random.random()<MUT_RATE:
            child=mutate(child, CHARS)
        next_gen.append(child)

    pop=next_gen
```

### <font color=green> Explicação do código 
    
<font color=green>O algoritmo começa criando uma população de 100 indivíduos, onde cada um é uma string aleatória com os mesmos caracteres possíveis da senha (letras e números). Cada indivíduo é avaliado por uma função de fitness, que compara o quão parecido ele está da senha-alvo. Quanto mais parecido, melhor a pontuação dele.
Depois disso, o algoritmo entra num loop que simula gerações. A cada geração, ele ordena os indivíduos pela pontuação de fitness e guarda os 10 melhores. Se algum deles for exatamente igual à senha, o programa para e mostra a geração em que isso aconteceu. Se não, ele continua cruzando os melhores indivíduos para formar novos filhos, e cada filho tem chance de sofrer mutação, ou seja, trocar algum caractere aleatoriamente. Isso ajuda a manter a diversidade e evitar que o algoritmo fique preso em combinações ruins.
No fim das contas, esse código mostra como dá pra usar conceitos de evolução e genética para resolver problemas de busca, como descobrir uma senha. Ele usa seleção dos melhores, cruzamento entre eles e mutações para ir melhorando a população com o tempo.
